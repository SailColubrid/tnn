```python
import keras
import numpy as np
from keras.utils.np_utils import to_categorical

X_train = np.load('data/train_x.npy')
y_train = np.load('data/train_y.npy')
X_test = np.load('data/test_x.npy')
y_test = np.load('data/test_y.npy')
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import SGD, Adam

x = Input(shape=(100,))
a = Dense(1024,activation='relu', kernel_initializer='VarianceScaling')(x)
b = Dense(10, activation='softmax')(a)
model = Model(inputs=x, outputs=b)
sgd = Adam()
model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x=X_train, 
          y=y_train, 
          batch_size=256, 
          epochs=200, 
          verbose=1, 
          validation_data=(X_test, y_test), 
          shuffle=True)
```
